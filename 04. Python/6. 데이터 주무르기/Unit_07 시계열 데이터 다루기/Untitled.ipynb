```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from fbprophet import Prophet
from datetime import datetime
```

```python
import platform

plt.rcParams['axes.unicode_minus'] = False
```

```python
pinkwink_web = pd.read_csv('08. PinkWink Web Traffic.csv', 
                                          encoding='utf-8', thousands=',',
                                          names = ['date','hit'], index_col=0)
pinkwink_web = pinkwink_web[pinkwink_web['hit'].notnull()]
pinkwink_web.head()
```

```python
pinkwink_web['hit'].plot(figsize=(12,4), grid=True);
```

```python
time = np.arange(0,len(pinkwink_web))
traffic = pinkwink_web['hit'].values

fx = np.linspace(0, time[-1], 1000)
```

```python
def mse(f, x, y):
    return np.sqrt(np.mean((f(x)-y)**2))
# mse - Mean Square Error
```

```python
fp1 = np.polyfit(time, traffic, 1) # 1차식
f1 = np.poly1d(fp1)

f2p = np.polyfit(time, traffic, 2) # 2차식
f2 = np.poly1d(f2p)

f3p = np.polyfit(time, traffic, 3) # 3차식
f3 = np.poly1d(f3p)

f15p = np.polyfit(time, traffic, 15) # 15차식
f15 = np.poly1d(f15p)

print(mse(f1, time, traffic))
print(mse(f2, time, traffic))
print(mse(f3, time, traffic))
print(mse(f15, time, traffic))
# error값이 점점 작아진다 - 좋아진다.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time, traffic, s=10)

plt.plot(fx, f1(fx), lw=4, label='f1')
plt.plot(fx, f2(fx), lw=4, label='f2')
plt.plot(fx, f3(fx), lw=4, label='f3')
plt.plot(fx, f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')

plt.legend(loc=2)
plt.show()
# 그렇게 큰 차이가 없다면 1차식 사용하는 것이 가장 이상적이다.
```

# 7-2. Prophet 모듈을 이용한 forecast 예측

```python
df = pd.DataFrame({'ds':pinkwink_web.index, 'y':pinkwink_web['hit']})
df.reset_index(inplace=True)
df['ds'] =  pd.to_datetime(df['ds'], format="%y. %m. %d.")
del df['date']
df.head()
```

```python
# prophet - 예측 ( 연단위의 주기성을 갖고있느냐 = yes)
m = Prophet(yearly_seasonality=True, daily_seasonality=True) 
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=60) # 60일간의 기간.
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
# 년, 주, 계절, 
```
