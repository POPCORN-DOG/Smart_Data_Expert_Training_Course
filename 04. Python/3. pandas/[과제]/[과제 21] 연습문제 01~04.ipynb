# <span style="color:deeppink"> 연습문제 3</span>

```python
# 다음 데이터프레임에서 지정하는 데이터를 뽑아내거나 처리하라.

data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
print(df,end='\n\n')

# 1. 모든 학생의 수학 점수를 시리즈로 나타낸다.
print(df['수학'],end='\n\n')            

# 2. 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
print(df[['국어','영어']],end='\n\n')

# 3. 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
df['평균점수'] = ((df['국어']+df['영어']+df['수학'])/3).round(2)
print(df,end='\n\n')

# 4. 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.
df.loc['방자','영어'] = 80
df['평균점수'] = ((df['국어']+df['영어']+df['수학'])/3).round(2)
print(df,end='\n\n')

# 5. 춘향의 점수를 데이터프레임으로 나타낸다.
print(df[:1],end='\n\n')

# 6. 향단의 점수를 시리즈로 나타낸다.
print(df.T['향단'])
```

# <span style="color:deeppink"> 연습문제 4</span>

```python
# 연습문제 4 (데이터프레임 고급 인덱싱)
# 1. 모든 행과 열에 라벨을 가지는 5x5 이상의 크기를 가지는 데이터 프레임을 만든다.
# 2. 10가지 이상의 방법으로 특정한 행과 열을 선택한다.

df = pd.DataFrame(np.arange(25).reshape(5, 5),
                   index=["가","나","다","라","마"],
                   columns=["A", "B", "C", "D","E"])

print(df,end='\n\n')

# 1. 행 1개 가져오기 (loc인덱서)
print(df.loc["가"],end='\n\n')   # 시리즈
print(df.loc[["가"]],end='\n\n') # 데이터프레임

# 2. 행 1개 이상 가져오기
print(df['가':'나'],end='\n\n')
print(df.loc['가':'나'],end='\n\n')
print(df.loc[['가','나']],end='\n\n')

# 3. 부등호로 가져오기
print(df.A > 10,end='\n\n')
print(df.loc[df.A >10],end='\n\n')

# 4. 행과 열을 모두 받는 경우 (데이터 1개)
print(df.loc['마','D'],end='\n\n')

# 5. 행과 열을 모두 받는 경우 (데이터 1개 이상) [행 : 열]
print(df.loc['가':,'A'],end='\n\n') # 시리즈이면서 A열을 가져올 수 있다.
print(df.loc['가':,'B'],end='\n\n') # B열
print(df.loc['가','A':],end='\n\n') # '가'행

# 6. 행과 열 잘라서 가져오기 df.loc[[행,행],[열,열]]
print(df.loc[['가','나'],['A','B']],end='\n\n')

# 7. 불리언 시리즈 활용해서 행과 열 잘라서 가져오기 df.loc[df.열>숫자,[열,열]]
print(df.loc[df.A>10,['A','B']],end='\n\n')

# =================================================================================
# =================================================================================
# 숫자는 항상 주의하자

# 8. iloc (라벨이 아니라 순서를 나타내는 정수 인덱스만 받는다)
print(df.iloc[0,1],end='\n\n')   # [행,열]
print(df.iloc[0:5,0],end='\n\n') # 시리즈가 된다. [0:전의 행까지]
print(df.iloc[0,-2:],end='\n\n') # 첫행의 끝에서 2번째까지

# 9. 행과 열 잘라서 가져오기 df.iloc 
print(df.iloc[2:3,1:3],end='\n\n') # df.iloc[행,열] 0부터 자기자신은 제외

# 10. 마이너스 부호(-)로 가져오기
print(df.iloc[-1],end='\n\n') # 시리즈
print(df.iloc[-2:],end='\n\n')

print(df,end='\n\n')
df.iloc[-1] = df.iloc[-1] * 5
print(df,end='\n\n')
```

# <span style="color:deeppink"> 연습문제 5</span>

```python
# 다음 명령으로 타이타닉호 승객 데이터를 데이터프레임으로 읽어온다. 
# 이 명령을 실행하려면 seaborn 패키지가 설치되어 있어야 한다.
# 타이타닉호 승객 데이터의 데이터 값을 각 열마다 구해본다.

import seaborn as sns

titanic = sns.load_dataset("titanic")

titanic.count()
```

# <span style="color:deeppink"> 연습문제 6</span>

```python
# 타이타닉호 승객중 

# 1. 성별(sex) 인원수
print(titanic.sex.value_counts(),end='\n\n')

# 2. 선실별(class) 인원수
print(titanic.pclass.value_counts().sort_index(),end='\n\n')

# 3. 사망/생존(alive) 인원수를 구하라.
print(titanic.alive.value_counts(),end='\n\n')

# 4. 나이별(age) 인원수
print(titanic.age.value_counts().sort_index(),end='\n\n')
```

# <span style="color:deeppink"> 연습문제 7</span>

```python
# 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고 각 그룹에 대해 
# '미성년자', '청년', '중년', '장년', '노년' 승객의 비율을 구한다. 
# 각 그룹 별로 비율의 전체 합은 1이 되어야 한다.

alive1 = titanic[titanic.alive == 'yes']
death1 = titanic[titanic.alive == 'no']

bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats1= pd.cut(alive1.age,bins,labels=labels)
cats2= pd.cut(death1.age,bins,labels=labels)

alive2 = cats1.value_counts()
print(alive2,end='\n\n')
print((alive2/sum(alive2)).round(2),end='\n\n')

death2 = cats2.value_counts()
print(death2,end='\n\n')
print((death2/sum(death2)).round(2),end='\n\n')
```
