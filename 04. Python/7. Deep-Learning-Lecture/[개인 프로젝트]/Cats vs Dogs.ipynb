```python
from keras.preprocessing import image
from keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.core import Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.utils import np_utils
from keras.callbacks import ModelCheckpoint, EarlyStopping
from glob import glob

import cv2, os, random
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = 'C:/Users/709-000/Desktop/dogs-vs-cats/train/' #바탕화면
```

```python
# 모델에서 사용할 이미지 크기 (96x96)

ROW, COL = 96, 96

dogs, cats = [], []
y_dogs, y_cats = [], []
```

```python
dog_path = os.path.join(path,'dog.5*')
len(glob(dog_path))
```

```python
## 강아지 이미지 처리 (1,111개 이미지)

dog_path = os.path.join(path, 'dog.5*') # 5번대 강아지 이미지 읽기
for dog_img in glob(dog_path):
    dog = cv2.imread(dog_img)                  # 이미지 읽기
    dog = cv2.cvtColor(dog,cv2.COLOR_BGR2GRAY) # 색상을 흑백으로 변경
    dog = cv2.resize(dog, (ROW, COL))          # 96x96으로 사진크기 변경
    dog = image.img_to_array(dog)              # 이미지를 배열 형태로 변경
    dogs.append(dog)
print('Some dog images starting with 5 loaded')
```

```python
## 고양이 이미지 처리 (1,111 개 이미지)
## 위와 같은 방식 

cat_path = os.path.join(path, 'cat.5*')
for cat_img in glob(cat_path):
    cat = cv2.imread(cat_img)
    cat = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)
    cat = cv2.resize(cat, (ROW, COL))
    cat = image.img_to_array(cat)
    cats.append(cat)
print('Some cat images starting with 5 loaded')
```

```python
classes = ['dog','cat']
```

```python
## 처리된 강아지 사진 확인

plt.figure(figsize=(12,8)) # figure 객체 사이즈

for i in range(5):
    plt.subplot(1,5,i+1)                               # 1x5의 이미지 출력 공간
    img = image.array_to_img(random.choice(dogs))      # 배열을 이미지로 변환
    plt.imshow(img,cmap=plt.get_cmap('gray'))          # 흑백으로 보이기
    
    plt.axis('off')                                    # 축 생략
    plt.title('It should be a {}.'.format(classes[0])) # 타이틀
plt.show()
```

```python
## 처리된 고양이 사진 확인 
## 위와 같은 방법 

plt.figure(figsize=(12,8))

for i in range(5):
    plt.subplot(1, 5, i+1)
    img = image.array_to_img(random.choice(cats))
    plt.imshow(img, cmap=plt.get_cmap('gray'))

    plt.axis('off')
    plt.title('It should be a {}.'.format(classes[1]))
plt.show()
```

```python
## 강아지의 라벨을 1로, 고양이의 라벨을 0으로 처리

y_dogs = [1 for item in enumerate(dogs)]
y_cats = [0 for item in enumerate(cats)]
```

```python
## converting everything to Numpy array to fit in our model
## them creating a X and target file like we used to see
## in Machine and Deep Learning models
## np.asarray()는 np.array()와는 달리 이미 ndarray가 있다면 복사(copy)를 하지 않는다.

dogs = np.asarray(dogs).astype('float32')
cats = np.asarray(cats).astype('float32')
y_dogs = np.asarray(y_dogs).astype('int32')
y_cats = np.asarray(y_cats).astype('int32')
```

```python
## 각 픽셀을 0~1값으로 변환 
dogs /= 255
cats /= 255
```

```python
X = np.concatenate((dogs,cats), axis=0)      # dogs 배열과 cats 배열을 합친다. 기준은 행으로
y = np.concatenate((y_dogs, y_cats), axis=0) # y_dogs 배열과 y_cats 배열을 합친다.
```

```python
len(X)  ## 강아지와 고양이 사진의 배열 수
```

```python
IMG_CHANNEL = 1           # image channel = 1 (흑백)
BATCH_SIZE = 32           # batch_size = 32
N_EPOCH = 40              # epoch = 40
VERBOSE = 2               # verbose = 2 
VALIDAION_SPLIT = .2      # 테스트셋 비율 20%
OPTIM = Adam()            # 최적화 함수 'adam'
N_CLASSES = len(classes)  # 클래스 수 = 2
```

```python
## 원-핫 인코딩 실행 

y = np_utils.to_categorical(y, N_CLASSES)
print('One-Hot Encoding done')
```

```python

## CNN 모델 생성 

model = Sequential([
    Conv2D(32, (3,3), padding='same', input_shape=(ROW, COL, IMG_CHANNEL), activation='relu'),
    Conv2D(32, (3,3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.25),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    Conv2D(64, (3,3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(.25),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(.5),
    Dense(N_CLASSES, activation='softmax')
])

print('The model was created by following config:')
model.summary()
```

```python
# 모델 컴파일 

model.compile(loss='categorical_crossentropy', 
                        optimizer=OPTIM, 
                        metrics=['accuracy'])
```

```python
# 모델 최적화 설정 및 저장 

MODEL_DIR = './CNN_Cats_vs_dogs_model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)

modelpath="./CNN_Cats_vs_dogs_model/{epoch:02d}-{val_loss:.4f}.hdf5"
    
checkpoint = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10)
```

```python
history = model.fit(X, y, batch_size=BATCH_SIZE, epochs=N_EPOCH, validation_split=VALIDAION_SPLIT,
          verbose=VERBOSE, callbacks=[early_stopping_callback, checkpoint])
```

```python
# 모델 정확도 

scores = model.evaluate(X, y, verbose=2)
print('MODEL ACCURACY\n{}: {}%'.format(model.metrics_names[1], scores[1]*100))
```

```python
# 테스트 셋의 오차
y_vloss = history.history['val_loss']

# 학습셋의 오차
y_loss = history.history['loss']

# 그래프로 표현
x_len = np.arange(1, len(y_loss)+1)
plt.plot(x_len, y_vloss, marker='.', c="red", label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset_loss')

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
