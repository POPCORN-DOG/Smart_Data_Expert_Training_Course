```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def sig10(x):
    return 1/(1 + np.exp(-10 * x))
def sig05(x):
    return 1/(1 + np.exp(-0.5 * x))
```

```python
X = np.linspace(-5,5,1001)
```

```python
sig_Y   = sigmoid(X)
sig_10Y = sig10(X)
sig_05Y = sig05(X)
```

```python
plt.plot(X, sig_Y,color='b')
plt.plot(X, sig_10Y,color='r') # 숫자가 커지면 가파라진다.
plt.plot(X, sig_05Y,color='g') # 숫자가 작으면 완만해진다.

plt.grid()
plt.show()
```

```python
def sig_shift(x, b):
    return 1/(1 + np.exp(-(x - b)))
```

```python
sig_shift2_Y = sig_shift(X, 2)
sig_shift2__Y = sig_shift(X, -2)
```

```python
plt.plot(X, sig_Y, color='b')
plt.plot(X, sig_shift2_Y, color='r')
plt.plot(X, sig_shift2__Y, color='g')
plt.grid()
plt.show()
```

```python
def derivative_sigmoid(x):
    return sigmoid(x) * (1-sigmoid(x)) # 미분
```

```python
derivative_Y = derivative_sigmoid(X) # 미분값
```

```python
plt.plot(X, sig_Y, color='b')
plt.plot(X, derivative_Y, color='r')
plt.grid()
plt.show()
```
